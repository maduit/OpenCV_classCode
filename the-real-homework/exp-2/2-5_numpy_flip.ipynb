```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img= cv2.imread('./test.jpg')
plt.imshow(img)
```

```python
img2 = img[:, :, ::-1]
plt.imshow(img2)
```

```python
# 一行代码完成垂直翻转
img3=cv2.flip(img2,0)
plt.imshow(img3)
```

```python
# 一行代码完成水平翻转
img4=cv2.flip(img2,1)
plt.imshow(img4)
```
